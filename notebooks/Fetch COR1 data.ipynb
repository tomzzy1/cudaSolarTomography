```python
import sunpy.map
from sunpy.net import Fido, attrs as a
```

```python
time = a.Time('2008-02-01 00:00:00', '2008-02-01 02:00:00')
instrument = a.Instrument.secchi
source = a.Source('STEREO_A')
```

```python
cor1a_result = Fido.search(time, instrument, source)
```

```python
cor1a_result.keys()
```

```python
for row in cor1a_result['vso']:
    if row['Extent Type'] == 'CORONA':
        print(row['fileid'])
```

```python
fts = Fido.fetch(cor1a_result['vso'][0])
```

```python
fts_map = sunpy.map.Map(fts)
```
